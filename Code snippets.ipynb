### The pretty version

```python
import spacy

nlp = spacy.load("en_core_web_lg")

doc = nlp(text)

spacy.displacy.render(doc, style='ent')
```

### The ugly but more functional version

```python
for ent in doc.ents:
    if ent.label_ == 'PERSON' or ent.label_ == 'ORG':
        print(ent.text)
```
